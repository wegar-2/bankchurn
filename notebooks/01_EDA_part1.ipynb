# 3. Credit Score

```python
cs_data = data[["credit_score", "churn"]].copy(deep=True)
```

```python
cs_data
```

```python
cs_data["credit_score"].hist(bins=100)
```

```python
cs_data["credit_score"].describe()
```

```python
cs_discretizer = KBinsDiscretizer(n_bins=10, strategy="quantile", encode="ordinal")
```

```python
cs_data.head()
```

```python
cs_data["credit_score_bin"] = cs_discretizer.fit_transform(cs_data["credit_score"].values.reshape(-1, 1))
```

```python
cs_data
```

```python
cs_ctab = pd.crosstab(index=cs_data["churn"], columns=cs_data["credit_score_bin"])
```

```python
cs_ctab
```

```python
cs_ctab.columns
```

# 6. Balance

```python
bal_data = data[["balance", "churn"]].copy(deep=True)
```

```python
bal_data["balance_bin"] = KBinsDiscretizer(n_bins=10, strategy="quantile", encode="ordinal").fit_transform(
    bal_data["balance"].to_numpy().reshape(-1, 1)
)
```

```python
bal_data.head()
```

```python
bal_ctab = pd.crosstab(index=bal_data["churn"], columns=bal_data["balance_bin"], normalize=True)
```

```python
bal_ctab
```

```python
bal_ctab.columns
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(1, 1, 1)
ax.bar(bal_ctab.columns, bal_ctab.loc[1, :])
ax.set_title("Churn share by Bin of balance", size=20)
ax.grid()
```

# 7. Analysis of Balance vs Age

I start by looking at the variables without excluding the cases for which balance is zero. This will be done later on.

```python
ba_data = data[["balance", "age", "gender"]].copy(deep=True)
```

```python
ba_data.head()
```

```python
ba_fig = plt.figure(figsize=(10, 7))
ba_ax = ba_fig.add_subplot(1, 1, 1)
ba_ax.scatter(y=ba_data["balance"], x=ba_data["age"])
```

The cases with zero balance are visible in the plot... excluding them now.

```python
ba_data = ba_data.loc[ba_data["balance"] > 0, :]
```

```python
ba_data
```

```python
ba_data["gender"].dtype
```

```python
ba_fig = plt.figure(figsize=(10, 7))
ba_ax = ba_fig.add_subplot(1, 1, 1)
ba_data["gender_color"] = ba_data["gender"].apply(lambda x: 1 if x == "M" else 0)
ba_ax.scatter(y=ba_data["balance"], x=ba_data["age"], c=ba_data["gender_color"])
ba_ax.legend()
ba_ax.grid()
```
